# **Object Detection Model**

*Description-*

**Goal:**
Develop an object recognition  model .The objective is to train a model that will be able to identify relevant Elements in a webpage.

**Objective:**

*   Implementing a computer vision architecture to perform element detection and identification.
*  
The model identifies elements like text,heading,buttons, logo images,Navigation dots, links, input fields etc.

*   Model creates a bounding box around each element and tag it with the desired element name.

**Dataset:**
Roboflow Dataset consisting of Website screenshots.

Train set-> 3382 Images and labels

Test set-> 482 Images and labels

Valid set-> 882 Images and labels

Data.yaml file



## Installing the dependencies and importing libraries

```python
!pip install ultralytics
import os
import cv2
from ultralytics import YOLO
from PIL import Image
from matplotlib import image as img
from matplotlib import pyplot as plt
```

### Training Dataset

```python
directory_path = '/content/drive/MyDrive/dataset/train/images'

# Getting a list of image files in the directory
image_files = [f for f in os.listdir(directory_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

for i in range(min(5, len(image_files))):
    image_path = os.path.join(directory_path, image_files[i])
    graph = cv2.imread(image_path)
    graph_rgb = cv2.cvtColor(graph, cv2.COLOR_BGR2RGB)

    # Displaying the image
    plt.imshow(graph_rgb)
    plt.title('Image {}'.format(i+1))
    plt.axis('off')
    plt.show()
```

```python
# Loading pretrained YOLO model
model = YOLO("yolov8n.pt")
```

## Training The Model

```python
model.train(data="/content/drive/MyDrive/dataset/data.yaml", epochs=3,batch=8,save=True)
```

```python
# Displaying the PR_curve as a result of training
PR_curve = cv2.imread('runs/detect/train/PR_curve.png')
plt.imshow(cv2.cvtColor(PR_curve, cv2.COLOR_BGR2RGB))
plt.title('PR_curve')
plt.axis('off')
plt.show()
```

```python
# Displaying results.png file as a result of training
Results = cv2.imread('runs/detect/train/results.png')
plt.imshow(cv2.cvtColor(Results, cv2.COLOR_BGR2RGB))
plt.title('Results')
plt.axis('off')
plt.show()
```

## Evaluation and Prediction

```python
model.val() # Shows performance scores and other metrics
```

```python
# Predictions

image_path="/content"
file_name="figma_webpage_1.png"

im1 = Image.open(os.path.join(image_path,file_name))
image = cv2.imread(os.path.join(image_path,file_name))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Image')
plt.axis('off')
plt.show()

 # saving plotted images
results = model.predict(source=im1, save=True)
saved_image=results[0].save_dir
saved_image_path=os.path.join(saved_image,file_name)

saved_image = cv2.imread(saved_image_path)
plt.imshow(cv2.cvtColor(saved_image, cv2.COLOR_BGR2RGB))
plt.title('YOLO Object Detection Result')
plt.axis('off')
plt.show()

# saving predictions as labels
im2 = cv2.imread(os.path.join(image_path,file_name))
results = model.predict(source=im2, save_txt=True)
```
